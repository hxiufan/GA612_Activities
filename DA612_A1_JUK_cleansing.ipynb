```python
import pandas as pd
import numpy as np
```

```python
# read the dataset
df = pd.read_csv('juk_meals.csv')
```

```python
df.head(10)
```

```python
df.dtypes
```

```python
# 1. Identify and handle missing values
missing_data = df.isnull().sum()

print("Missing data per column:")
print(missing_data)
```

```python
# Identify the duplicated data
df.duplicated().sum()
```

```python
# 2. Change the column "Unnamed:0" to Product ID
df.rename(columns={'Unnamed: 0': 'Product ID'}, inplace=True)
```

```python
# 3. Rename the product ID following the rule of JUK0001, JUK0002
df["Product ID"] = ["JUK{:04d}".format(i + 1) for i in range(len(df))]
```

```python
df
```

```python
# 4. Remove "From" and "\n" in the column "Price"
df['Price'] = df['Price'].str.replace("From ", "").str.replace("\n", "", regex=False)
```

```python
df
```

```python
# 5. Add a column "Regular Price" on the left of "Price"
df.insert(2, "Regular Price", 0.0)
```

```python
# 6. Rename the "Price" column as "Price Range"
df.rename(columns={"Price": "Price Range"}, inplace=True)
```

```python
#7. Fill the "Regular Price" with the first price form "Price Range", id the first six characters
df["Regular Price"] = df["Price Range"].str[:6]
```

```python
# 8. Convert the data type of "Regular Price" column to float with 2 decimals
df["Regular Price"] = df["Regular Price"].str.replace("$", "").astype(float).round(2)
```

```python
#9. Replace the URL with each product's full URL
df["URL"] = "https://www.juk.co.nz/collections/all-meals" + df["URL"]
```

```python
# 10. Locate the row with " "Paneton Sourdough Ficelle – Twin Pack", as it is the first row of unrelevant content
target_row = df[df["Meal"] == "Paneton Sourdough Ficelle – Twin Pack"]
target_row
```

```python
# List the data from row 27 to the ened, to review if they are all unrelevant to frozen meals
df.iloc[27:]
```

```python
# Delete the rows from row 27 onward
df = df.drop(df.index[27:])
```
