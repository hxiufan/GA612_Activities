```python
titles = []
prices = []
urls = []
images = []

for i in range (1,3):
  url = f"https://www.juk.co.nz/collections/all-meals?page={i}"

  r = requests.get(url)
  content = r.content
  soup = BeautifulSoup(content,"html.parser")

  # Find all meals
  main_table = soup.find('div', class_='collection span-12')


# Extract meals titles and URLs
  meal_titles = main_table.find_all('p', class_='product-loop__title h4 mb2 sm-mb1')
  for title in meal_titles:
    meal_title = title.find('a').text.strip()
    meal_url = title.find('a')['href']
    titles.append(meal_title)
    urls.append(meal_url)

    # Extract meal images
  meal_images = main_table.find_all('div', class_='product-loop__img-wrapper relative')
  for image in meal_images:
    meal_image = image.find('a')['href']
    images.append(meal_image)

# Extract meal prices
  meal_prices = main_table.find_all('div', class_='product-loop__price mt0 pb1 px1 sm-px0 a-center')
  for price in meal_prices:
    meal_price = price.text.strip().replace('Unit price\n/per', '').strip() # removed the empty field of Unit price
    prices.append(meal_price)



```

```python
data_juk = {'Meal': titles, 'Price': prices, 'URL':urls, 'Image':images}
df = pd.DataFrame(data_juk)
```
