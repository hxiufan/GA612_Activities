```python
# Start the data cleaning for 'BL-Flickr-Images-Book.csv'
df2 = pd.read_csv('BL-Flickr-Images-Book.csv')

# To display the first few rows of the DataFrame
df2.head()
```

```python
# Identify and handle missing values
missing_data = df2.isnull().sum()

print("Missing data per column:")
print(missing_data)
```

```python
df2.isnull()
```

```python
df2.isna().sum()
```

```python
# Drop the columns with a significant number of missing values, in this case they are not key to this analysis.
to_drop = ['Edition Statement',
            'Corporate Author',
            'Corporate Contributors',
            'Former owner',
            'Engraver']

df2.drop(to_drop,inplace=True, axis=1)
```

```python
df2.head()
```

```python
# Check if any duplicated variables in "Indentifier"
df2['Identifier'].is_unique
```

```python
df2 = df2.set_index('Identifier')
df2.head()
```

```python
df2['Identifier'] = df2.index
df2 = df2.set_index('Identifier')
df2.head()
```

```python
df2.loc[206]
```

```python
df2.dtypes.value_counts()
```

```python
df2.loc[1905:, 'Date of Publication'].head(10)
```

```python
# Remove the extra dates in square brackets, wherever present: 1879 [1878]
# Convert date ranges to their “start date”, wherever present: 1860-63; 1839, 38-54
# Completely remove the dates we are not certain about and replace them with NumPy’s NaN: [1897?]
# Convert the string nan to NumPy’s NaN value
```

```python
#find any four digit at the beginning of a string,\d means any digit, {4} meains repeats this rule for four times
regex = r'^(\d{4})'
```

```python
extr = df2['Date of Publication'].str.extract(r'^(\d{4})', expand=False)
```

```python
extr.head()
```

```python
df2['Date of Publication'] = pd.to_numeric(extr)
df2['Date of Publication'].dtype
```

```python
df2['Date of Publication'].isnull().sum()/len(df2)
```

```python
# Cleaning "Place of Publication" field
```

```python
# Step 1: Show the content of the column
df2['Place of Publication'].head(10)
```

```python
pub = df2['Place of Publication']
london = pub.str.contains('London')
london[:5]
```

```python
oxford = pub.str.contains('Oxford')
```

```python
df2['Place of Publication'] = np.where(london, 'London',
                                       np.where(oxford, 'Oxford',
                                                pub.str.replace('-',' ')))
df2['Place of Publication'].head()
```

```python
df2.head()
```

```python
# Replace the missing value in 'Author' with a placeholder 'Unknown'
df2['Author'].fillna("Unknown", inplace = True)
```

```python
# Verify the changes
df2.isna().sum()
```

```python
# Replace the missing value in 'Publisher' with a placeholder 'Unknown'
df2['Publisher'].fillna("Unknown", inplace = True)
```

```python
df2.isna().sum()
```

```python
# Imputing the missing value in 'Date of Publication' with mode
df2['Date of Publication'].describe(), df2['Date of Publication'].mode()
```

```python
df2['Date of Publication'].fillna(df2['Date of Publication'].mode()[0], inplace=True)
```
