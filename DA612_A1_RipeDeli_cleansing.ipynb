```python
# Check spelling with textblob, but not ideal so terminated this part
from textblob import TextBlob
```

```python
def correct_spelling(text):
    return str(TextBlob(text).correct())
```

```python
df["Corrected Meal"] = df["Meal"].apply(correct_spelling)
```

```python
print(df[["Meal", "Corrected Meal"]])
```

Data cleansing:

```python
# 1. Rename the first row, adding company's initial as FC+ 4 digit numbers, from 0001
df.rename(columns={'Unnamed: 0': 'Product ID'}, inplace=True)
df["Product ID"] = ["RD{:04d}".format(i + 1) for i in range(len(df))]
```

```python
# 2.Idnetify if all the price are with more options, i.e. to check if all prices are with "from"
df['Price'].str.contains('from').all()
```

```python
# Identify the row(s) without 'from'
df[~df['Price'].str.contains('from')]
```

```python
# 3. Add a column next to 'Price', named 'Larger Size Available'.
df.insert(3, "Larger Size Available", False)
```

```python
df["Larger Size Available"] = df["Price"].str.contains("from")
```

```python
# 4. Conver the data type of "Price" to numeric type
df['Price'] = df['Price'].str.replace('from', '')
```

```python
df["Price"] = df["Price"].str.replace("$", "").astype(float).round(2)
```

```python
# Clean the 'Price' column, handle the value of 'Sold Out', and replace them with a median value of the column
```

```python
sold_out_rows = df[df['Price'] == 'Sold Out']
```

```python
sold_out_rows
```

```python
# Convert the data type to numeric in all the price with '$'
def clean_price(price):
    if price == 'Sold Out':
        return None
    else:
        return float(price.replace('$', ''))
```

```python
df['Numeric_Price'] = df['Price'].apply(clean_price)
```

```python
median_price = df['Numeric_Price'].median()
```

```python
median_price
```

```python
# Replace 'Sold Out' with the median price in the 'Numeric_Price' column
df['Numeric_Price'] = df['Numeric_Price'].fillna(median_price)
```

```python
df['Numeric_Price']
```

```python
df
```

```python
# Replace the values in 'Price' with the values from 'Numeric_Price'
df['Price'] = df['Numeric_Price']
```

```python
# Drop the 'Numeric_Price' column
df.drop(columns=['Numeric_Price'], inplace=True)
```

```python
#5. Replace the URL with each product's full URL
df["URL"] = "https://www.ripedeli.co.nz/collections" + df["URL"]
```

```python
# Tidy up the URLs in image, remove "//"
df["Image"] = df["Image"].str.replace("//", "")
```
