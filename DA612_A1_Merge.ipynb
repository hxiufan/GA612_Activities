```python
import pandas as pd
```

```python
# Read the datasets
fc_cleaned = pd.read_csv('/content/drive/My Drive/DA612/fc_cleaned.csv')
juk_cleaned = pd.read_csv('/content/drive/My Drive/DA612/juk_cleaned.csv')
ripe_deli_cleaned = pd.read_csv('/content/drive/My Drive/DA612/RipeDeli_cleaned.csv')
```

```python
# Merge the dataframes into one, using outer
merged_df = pd.concat([fc_cleaned, juk_cleaned, ripe_deli_cleaned], join = 'outer')
```

```python
merged_df
```

```python
# Set "Product ID" as index, because it has unique value
merged_df.set_index('Product ID', inplace=True)
```

```python
merged_df
```

```python
# Reset the index to the default integer-based index
merged_df.reset_index(drop=False, inplace=True)
```

```python
merged_df
```

```python
# Generate the lists of desired Product IDs
fc_ids = [f'FC{str(i).zfill(4)}' for i in range(1, 11)]
rd_ids = [f'RD{str(i).zfill(4)}' for i in range(1, 11)]
juk_ids = [f'JUK{str(i).zfill(4)}' for i in range(1, 11)]
```

```python
top10_df = merged_df[merged_df['Product ID'].isin(fc_ids + rd_ids + juk_ids)][['Product ID','Meal','Price']]
```

```python
top10_df
```

```python
# Perform multi-level indexing by setting mulit columns as index
# Create a MultiIndex
fc_columns = ['Product ID', 'Meal', 'GF Option', 'Price', 'URL', 'Image']
juk_columns = ['Product ID', 'Meal', 'Price', 'Price Range', 'URL', 'Image']
ripe_deli_columns = ['Product ID', 'Meal', 'Price', 'Larger Size Available', 'URL', 'Image']

fc_df = fc_cleaned[fc_columns].set_index(['Product ID'])
fc_df.columns = pd.MultiIndex.from_product([['FreshChef'], fc_df.columns])

juk_df = juk_cleaned[juk_columns].set_index(['Product ID'])
juk_df.columns = pd.MultiIndex.from_product([['JUK'], juk_df.columns])

ripe_deli_df = ripe_deli_cleaned[ripe_deli_columns].set_index(['Product ID'])
ripe_deli_df.columns = pd.MultiIndex.from_product([['RipeDeli'], ripe_deli_df.columns])

# Concatenate the dataframes along columns
final_df = pd.concat([fc_df, juk_df, ripe_deli_df], axis=1)
```

```python
final_df
```

```python
# Sort the merged dataframe based on 'Price' in ascending order
sorted_df = merged_df.sort_values(by='Price', ascending=True)
```

```python
# Perform summary statistics on the 'Price' column of 'merged_df'
summary_statistics = merged_df['Price'].describe()

summary_statistics
```

```python
# Slice the dataframe to get rows where the 'Price' is greater than 20 and less than 40
sliced_df = merged_df[(merged_df['Price'] > 20) & (merged_df['Price'] < 40)]
```
