```python
import pandas as pd
```

```python
# Read the datasets
fc_cleaned = pd.read_csv('/content/drive/My Drive/DA612/fc_cleaned.csv')
juk_cleaned = pd.read_csv('/content/drive/My Drive/DA612/juk_cleaned.csv')
ripe_deli_cleaned = pd.read_csv('/content/drive/My Drive/DA612/RipeDeli_cleaned.csv')
```

```python
# Merge the dataframes into one, using outer
merged_df = pd.concat([fc_cleaned, juk_cleaned, ripe_deli_cleaned], join = 'outer')
```

```python
merged_df
```

```python
# Set "Product ID" as index, because it has unique value
merged_df.set_index('Product ID', inplace=True)
```

```python
merged_df
```

```python
# Reset the index to the default integer-based index
merged_df.reset_index(drop=False, inplace=True)
```

```python
merged_df
```

```python
# Generate the lists of desired Product IDs
fc_ids = [f'FC{str(i).zfill(4)}' for i in range(1, 11)]
rd_ids = [f'RD{str(i).zfill(4)}' for i in range(1, 11)]
juk_ids = [f'JUK{str(i).zfill(4)}' for i in range(1, 11)]
```

```python
top10_df = merged_df[merged_df['Product ID'].isin(fc_ids + rd_ids + juk_ids)][['Product ID','Meal','Price']]
```

```python
top10_df
```

```python
# Sort the merged dataframe based on 'Price' in ascending order
sorted_df = merged_df.sort_values(by='Price', ascending=True)
```
