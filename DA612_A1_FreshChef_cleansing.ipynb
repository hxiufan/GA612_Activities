```python
import pandas as pd
```

```python
# read the dataset
df = pd.read_csv('FreshChef_meals.csv')
```

```python
df.head(10)
```

```python
# Identify the data types in each column
df.dtypes
```

```python
# Identify and handle missing values
missing_data = df.isnull().sum()

print("Missing data per column:")
print(missing_data)
```

```python
# Identify duplicated values
df.duplicated().sum()
```

```python
# 1. Rename the first row, adding company's initial as FC+ 4 digit numbers, from 0001
df.rename(columns={'Unnamed: 0': 'Product ID'}, inplace=True)
df["Product ID"] = ["FC{:04d}".format(i + 1) for i in range(len(df))]
```

```python
# 2. Add a column "Regular Price" on the left of "Price"
df.insert(2, "GF Option", False)
```

```python
# 3. Update the "GF Option" column: if there is  "✓" in the "Meal" column, GF Option is True
df["GF Option"] = df["Meal"].str.contains("✓")
```

```python
# 4. Delte all the "✓" from the "Meal" column
df["Meal"] = df["Meal"].str.replace("✓", "").str.strip()
```

```python
#5.  Replace the URL with each product's full URL
df["URL"] = "https://www.ripedeli.co.nz/collections" + df["URL"]
```

```python
#6. Tidy up the URLs of the images
df['Image'] = df['Image'].str.replace('//', 'https://www.')
```
