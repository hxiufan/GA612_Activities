```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://freshchef.nz/collections/frontpage"
```

```python
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
table = soup.find('div', class_='product-grid-container scroll-trigger animate--slide-in')
```

```python
meals = table.find_all('li', class_='grid__item scroll-trigger animate--slide-in')
```

```python
meals_all = []
```

```python
for meal in meals:
   title = meal.find('h3').text.replace('\n', '')
   price = meal.find('span',class_='price-item price-item--regular').text.strip()
   price = float(price[1:])

   meals_all.append([title,price])

```

```python
print (meals_all)
```

```python
data = pd.DataFrame(meals_all, columns=['Meal', 'Price'])
```
